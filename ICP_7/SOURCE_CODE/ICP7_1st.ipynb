```
import requests
from bs4 import BeautifulSoup     #importing all requied libraries
url="https://en.wikipedia.org/wiki/Google"   #storing url into url variable
code=requests.get(url)         #making a get request
text=code.text
soup=BeautifulSoup(text,"html.parser")
file1 = open("/content/drive/My Drive/Colab Notebooks/webscrap.txt","w")
for l in soup.find_all('p'):                  #printing all urls without anchor tag
    file1.writelines(str(l.get_text(strip=True)))
```

```
#Tokenization
import nltk

nltk.download('punkt')
file2 =open("/content/drive/My Drive/Colab Notebooks/webscrap.txt","r")
w=""
w=file2.readline()
word_token=nltk.word_tokenize(w)
for r in word_token:
   print(r)
```

```
#POS
nltk.download('averaged_perceptron_tagger')
tagged=nltk.pos_tag(word_token)
print(tagged)
```

```
#Stemming
from nltk.stem import PorterStemmer
pStemmer=PorterStemmer()
for r in word_token:
   print(pStemmer.stem(r))
```

```
#Lemmatization
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lem= WordNetLemmatizer()

for r in word_token:
   print(lem.lemmatize(r))
```

```
#Trigram
from nltk.util import ngrams
trigrams=ngrams(word_token,3)
for t in trigrams:
  print(t)
```

```
#NAMED ENTITY RECOGNITION
from nltk import wordpunct_tokenize, pos_tag, ne_chunk
nltk.download('maxent_ne_chunker')
nltk.download('words')
print(ne_chunk(pos_tag(wordpunct_tokenize(w))))
```
