```
#Tokenization
import nltk

nltk.download('punkt')
file2 =open("/content/drive/My Drive/Colab Notebooks/nlp_input.txt","r",encoding="utf-8")
#file2.decode('UTF8','replace')
w=""
w=file2.read()

word_token=nltk.word_tokenize(w)
for r in word_token:
   print(r)
```

```
#Lemmatization
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lem= WordNetLemmatizer()

for r in word_token:
   print(lem.lemmatize(r))
```

```
#TRIGRAM
from nltk.util import ngrams
trigrams=ngrams(word_token,3)
l=[]
for t in trigrams:
  l.append(t)
  print(t)
```

```
#getting the trigrams with frequency
wordFrequency = nltk.FreqDist(l)
Frequency_Trigrams = wordFrequency.most_common()
print("TRIGRAMS WITH FREQUENCY:\n",Frequency_Trigrams)
```

```
# Top 10 Trigrams.

Top_ten = wordFrequency.most_common(10)

print("Top 10 Trigrams:\n", Top_ten)
```

```
sentence = nltk.sent_tokenize(w)

Final_Result = []

# Iterating the Sentences

for s in sentence:
    # Iterating through each sentence

  for a, b, c in l:
      
        #iterating through top 10 trigrams 

        for ((i, j, k), length) in wordFrequency.most_common(20):
          #iterating through all trigrams 

            if(a, b, c == i, j, k):

                Final_Result.append(s)



print("Final Result after concatination: \n", Final_Result)
```
